```python
def count_numbers(number_list):
    count_dict = {}

    for number in number_list:
        if number in count_dict:
            count_dict[number] += 1
        else:
            count_dict[number] = 1

    return count_dict

def filter_dicts(result_dict):
        final_dict2 = {}
        for key,value in result_dict.items():
            if value > 1:
               final_dict2[key] = value
        return final_dict2


numbers = [5, 5, 5, 6, 7, 7]
result_dict = count_numbers(numbers)
final_dict = filter_dicts(result_dict)
print(final_dict)
```
